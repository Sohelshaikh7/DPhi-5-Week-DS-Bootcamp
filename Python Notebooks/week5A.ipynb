```
# importing packages
import numpy as np # to perform calculations 
import pandas as pd # to read data
import matplotlib.pyplot as plt # to visualise
import seaborn as sns
```

```
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/kc_house_data/kc_house_data.csv')
```

```
data.head()
```

```
X = data.drop(['price'],axis=1)

y = data[['price']]
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```
from sklearn.linear_model import LinearRegression
```

```
lm = LinearRegression()
```

```
lm.fit(X_train,y_train)
```

```
lm_pred = lm.predict(X_train)
```

```
from sklearn import metrics
```

```
print('R^2:',metrics.r2_score(y_train, lm_pred))
```

```
from sklearn.tree import DecisionTreeRegressor
```

```
dt = DecisionTreeRegressor(random_state = 42)
```

```
dt.fit(X_train,y_train)
```

```
dt_pred = dt.predict(X_train)
```

```
print('R^2:',metrics.r2_score(y_train, dt_pred))
```

```
dt1 = DecisionTreeRegressor(max_depth = 3, random_state = 42)
```

```
dt1.fit(X_train,y_train)
```

```
dtt_pred = dt1.predict(X_train)
```

```
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, dtt_pred)))
```

```
dt1_pred = dt1.predict(X_test)
```

```
print('R^2:',metrics.r2_score(y_test, dt1_pred))
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
rf = RandomForestRegressor(random_state = 42)
```

```
rf.fit(X_train,y_train)
```

```
rf_pred = rf.predict(X_test)
```

```
print('MAE:', metrics.mean_absolute_error(y_test, rf_pred))
print('MSE:', metrics.mean_squared_error(y_test, rf_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rf_pred)))
print('R^2:',metrics.r2_score(y_test, rf_pred))
```

```
rf1 = RandomForestRegressor(n_estimators= 10,random_state = 42)
```

```
rf1.fit(X_train,y_train)
```

```
rf1_pred = rf1.predict(X_test)
```

```
print('MAE:', metrics.mean_absolute_error(y_test, rf1_pred))
print('MSE:', metrics.mean_squared_error(y_test, rf1_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rf1_pred)))
print('R^2:',metrics.r2_score(y_test, rf1_pred))
```
